```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
from matplotlib import style
import seaborn as sb
```

```python
df=pd.read_csv("C:\\Users\\Joovin\\OneDrive\\Documents\\houseprice1.csv")
df
```

```python
sb.barplot(x='location',y='price',data=df)
```

```python
data = df.groupby("location")["price"].sum()
data
```

```python
pie, ax = plt.subplots(figsize=[10,6])
labels = data.keys()
plt.pie(x=data, autopct="%.1f%%", labels=labels, pctdistance=0.5)
```
